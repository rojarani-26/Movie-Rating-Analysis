```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("/content/movie_dataset.csv")

print(df.head())
print(df.info())
print(df.isnull().sum())
```

```python
df = df.dropna(subset=['Rating'])

df['Budget'] = df['Budget'].fillna(0)
df['Revenue'] = df['Revenue'].fillna(0)

print(df)
```

```python
ratings = df['Rating'].values
budget = df['Budget'].values
revenue = df['Revenue'].values

print("Ratings Statistics")
print("Mean:", np.mean(ratings))
print("Median:", np.median(ratings))
print("Standard Deviation:", np.std(ratings))

print("Budget Statistics")
print("Mean:", np.mean(budget))
print("Median:", np.median(budget))
print("Standard Deviation:", np.std(budget))

print("Revenue Statistics")
print("Mean:", np.mean(revenue))
print("Median:", np.median(revenue))
print("Standard Deviation:", np.std(revenue))
```

```python
genre_avg = df.groupby('Genre')['Rating'].mean()
print("Average Rating by Genre:\n", genre_avg)

director_avg = df.groupby('Director')['Rating'].mean()
print("\nAverage Rating by Director:\n", director_avg)

year_avg = df.groupby('Year')['Rating'].mean()
print("\nAverage Rating by Year:\n", year_avg)
```

```python
genre_avg.plot(kind='bar',title='Average Movie Rating by Genre')
plt.xlabel('Genre')
plt.ylabel('Average Rating')
plt.show()
```

```python
director_avg.plot(kind='bar',title='Top Directors by Average Rating')
plt.xlabel('Director')
plt.ylabel('Average Rating')
plt.show()
```

```python
year_avg.plot(kind='line', marker='o',title='Average Movie Ratings Over Years')
plt.xlabel('Year')
plt.ylabel('Average Rating')
plt.grid(True)
plt.show()
```

```python
df.plot(
    kind='scatter',
    x='Budget',
    y='Rating',
    title='Budget vs Movie Rating'
)
plt.show()

```

```python
df.plot(
    kind='scatter',
    x='Revenue',
    y='Rating',
    figsize=(8,5),
    title='Revenue vs Movie Rating'
)
plt.show()
```

```python
top_movies = df.sort_values(by='Rating', ascending=False).head(10)

top_movies.plot(
    kind='barh',
    x='Title',
    y='Rating',
    title='Top 10 Highest Rated Movies'
)

plt.gca().invert_yaxis()
plt.show()

```
